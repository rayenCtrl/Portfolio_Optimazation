```python
# calculate sharpe ratio 

import numpy as np

def sharpe_ratio(returns, rf, days=252):
    volatility = returns.std() * np.sqrt(days)
    sharpe_ratio = (returns.mean() - rf) / volatility
    return sharpe_ratio


# calculate sharpe ratio for apple with risk free rate of 5%
sharpe_ratio(hist['Close'], 0.05)

print('%.2f' % sharpe_ratio(hist['Close'], 0.05))
```

```python
# can you create new df with only the close price and date
df=hist
df=df[['Close']]
df
```

```python
# clean up the date column 
df.reset_index(inplace=True)
df
```

```python
#can you keep only the date (Y-m-d) format 
df['Date']=df['Date'].dt.strftime('%Y-%m-%d')
df
```

```python
# can you rename the column close to y and date to ds
df=df.rename(columns={'Close':'y','Date':'ds'})
df
```

- Tool for producing high quality forecasts for time series data that has multiple seasonality with linear or non-linear growth.
https://github.com/facebook/prophet/tree/main/docs/_docs

```python
from prophet import Prophet
m=Prophet()
m.fit(df)
```

```python
future=m.make_future_dataframe(periods=365)
future.tail()
```

```python
forecast=m.predict(future)
forecast[['ds','yhat','yhat_lower','yhat_upper']].tail()
```

```python
fig1=m.plot(forecast)
```

```python
fig2=m.plot_components(forecast)
```
